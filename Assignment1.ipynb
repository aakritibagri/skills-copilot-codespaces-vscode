```python

import matplotlib.pyplot as plt
import math as m
a=[]
b=[]
#A
n=int(input("Enter the total no. of elements:"))
for i in range(0,n):
      a.append(int(input("Enter the values for A:")))
print(a)
sum=0
for i in range (0,n):
      sum=sum+a[i]
mean=sum/n
print("Mean value of A list:",mean)
sq=0
for i in range (0,n):
      sq=sq+(a[i]**2)
r=sq/n
sd=m.sqrt(r-(mean**2))
print("Standard Deviation  of A list:",sd)
print("Variance of A list:",sd**2)
#B
n=int(input("Enter the total no. of elements:"))
for i in range(0,n):
      b.append(int(input("Enter the values for B:")))
print(b)
add=0
for i in range (0,n):
      add=add+b[i]
mean1=add/n
print("Mean value of B list:",mean1)
c=0
for i in range (0,n):
      c=c+(b[i]**2)
q=c/n
sm=m.sqrt(q-(mean1**2))
print("Standard Deviation of B list:",sm)
print("Variance of B list:",sm**2)
#Correlation
r=0
for x in range (0,n):
  r=r+ (a[x]*b[x])
u=r/n
t=(u-(mean*mean1))
s=t/(sd*sm)
print("Correlation coefficient r:",s)
#Covariance
e=0
for x in range (0,n):
    temp=a[x]
for x in range (0,n):
  e=e+((a[x]-mean)*(b[x]-mean1))
cov=e/n
print("Covariance :",cov)
#Covariance matrix
t=[[(sd**2),cov],[cov,(sm**2)]]
print(t)
plt.plot(a)
plt.show()
```
