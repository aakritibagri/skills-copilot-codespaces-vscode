```python

import seaborn as sns
import matplotlib.pyplot as plt

# Assuming you have loaded your dataset into a DataFrame 'df'

# Create separate FacetGrids for each column
g1 = sns.FacetGrid(df, hue="Which platform you are most active ?", height=4, aspect=2)
g1.map(sns.distplot, "Social Media").add_legend()
plt.title("Distribution of Social Media Usage by Platform")

g2 = sns.FacetGrid(df, hue="Which platform you are most active ?", height=4, aspect=2)
g2.map(sns.distplot, "News apps").add_legend()
plt.title("Distribution of News Apps Usage by Platform")

g3 = sns.FacetGrid(df, hue="Which platform you are most active ?", height=4, aspect=2)
g3.map(sns.distplot, "Newspaper").add_legend()
plt.title("Distribution of Newspaper Usage by Platform")

# Show the plots
plt.show()
```
